```
df_ = df[df.Crop.isin(["Rice","Wheat","Maize","Ragi"])].groupby(["State_Name","Crop_Year"]).sum()
```

```
df_.head()
```

```
df_.reset_index(inplace=True)
```

```
df.head()
```

```
df_.sort_values("Crop_Year",inplace=True)
```

```
px.scatter(data_frame=df_,x="Area",y="Production",animation_frame="Crop_Year",animation_group="State_Name",color="State_Name")
```

```
df_["Efficiency"] = df_["Production"]/df_["Area"]
```

```
px.scatter(data_frame=df_,x="Area",y="Efficiency",size="Production",animation_frame="Crop_Year",animation_group="State_Name",color="State_Name",range_y=[0.75,5],range_x=[-1E6,20E6])
```
