```
df = pd.read_csv("apy.csv", na_values="=");
```

```
df.dropna(inplace=True)
```

```
df.isnull().sum()
```

```
sns.kdeplot(df.Production);
```

```
sns.boxplot(df.Production);
```

```

sns.boxplot(df.Area);
```

```
df[df.State_Name == "Karnataka"]["District_Name"].unique()
```

```
df.groupby(["State_Name","Crop","Crop_Year"]).sum()
```

```
df_ = df.groupby(["State_Name","Crop_Year"]).sum()
```

```
df_.reset_index(inplace = True)
```

```
df_.head()
```

```
df_[["State_Name","Crop_Year"]].groupby(["State_Name"]).count()
```

```
sns.lineplot(x="Crop_Year",y="Production",data=df_);
```

```
sns.lineplot(x="Crop_Year",y="Production",data=df_,estimator=np.median);
```

```
sns.lineplot(x="Crop_Year",y="Area",data=df_[df_["State_Name"] == "Andaman and Nicobar Islands"],estimator=np.median);
```

```
sns.lineplot(x="Crop_Year",y="Production",data=df_[df_["State_Name"] == "Andaman and Nicobar Islands"],estimator=np.median);
```

```
sns.lineplot(x="Crop_Year",y="Production",hue="State_Name",data=df_,estimator=np.median);
```

```
px.scatter(data_frame=df_,x="Area",y="Production",animation_frame="Crop_Year",animation_group="State_Name",color="State_Name")
```
